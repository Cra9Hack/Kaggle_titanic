```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier 
```

```python
work_df = pd.read_csv('train.csv')
raw_data = pd.read_csv('train.csv')
```

```python
work_df['Deck'] = [ x[0] if type(x) == str else np.NaN for x in work_df['Cabin'] ]
work_df = work_df.drop(['Cabin'], axis = 1)
work_df['Deck'] = work_df.groupby('Ticket')['Deck'].transform(lambda x: x.fillna( x.value_counts(dropna = False).index[0])  )
work_df['Group'] = [ x.split(',')[0] for x in raw_data['Name']]
work_df.loc[work_df['Group'].value_counts()[work_df['Group']].values > 1 , 'Group'] = "in_group"
work_df.loc[work_df['Group'].value_counts()[work_df['Group']].values ==1 , 'Group'] = "none"
print(work_df['Group'].value_counts()['none'])
work_df.loc[work_df['Ticket'].value_counts()[work_df['Ticket']].values > 1 , 'Group'] = "in_group"
print(work_df['Group'].value_counts()['none'])

work_df['Deck'] = work_df.groupby('Group')['Deck'].transform(lambda x: x.fillna( x.value_counts(dropna = False).index[0])  )
work_df['Deck'] = work_df.groupby('Fare')['Deck'].transform(lambda x: x.fillna( x.value_counts(dropna = False).index[0])  )
#work_df['Deck'] = work_df['Deck'].fillna('NO')
```

```python
work_df['Embarked'] = work_df['Embarked'].fillna('nan')
work_df.loc[work_df['Embarked']=='nan']
work_df['Embarked'] = work_df['Embarked'].replace('nan','S')
```

```python
work_df['Title'] = [ x.split(',')[1].split('.')[0][1:] for x in work_df['Name']]
def title(x):
    if x['Title'] in ['Mr','Mrs','Miss']:
        return 'titnorm'
    if x['Title'] in ['Master', 'Rev']:
        return 'titgood'
    else:
        return 'titbad'

work_df['Title_chance'] = work_df.apply(title, axis = 1)
```

```python
work_df['Age']  = work_df.groupby(["Sex","Pclass"])['Age'].transform(lambda x: x.fillna(x.median()))
```
