```python


svc = SVC()
svc_params = {'C': np.arange(10,200,10), 
              'gamma': np.arange(0.0005,0.003,0.0005),
              #'kernel': ['linear', 'poly', 'rbf', 'sigmoid'] 
             } 
svc_opt = GridSearchCV(estimator = svc, param_grid = svc_params, n_jobs = -1, cv=10 , scoring = 'f1')


knn = KNeighborsClassifier()
# knn_params = { 'n_neighbors' : np.arange(2,15,1),
#                'weights' : ['uniform','distance'],
#                'metric' : ['minkowski','euclidean','manhattan']}
knn_params = { 'n_neighbors' : np.arange(2,15,1)}
knn_opt = GridSearchCV(estimator = knn, param_grid = knn_params, n_jobs = -1, cv = 10,scoring = 'f1')


dt = DecisionTreeClassifier()
dt_params = {'min_samples_leaf': [1, 2, 3,4,5],
             'max_depth': [ 2,3,4, 5, 6],
             'class_weight': [None,'balanced']}

dt_opt = GridSearchCV(estimator = dt, param_grid = dt_params, n_jobs = -1, cv = 10,scoring = 'f1')


logreg = LogisticRegression(max_iter = 10000)
logreg_params =     {
    'penalty' : ['l1', 'l2'],
    'C' : np.logspace(-4, 4, 20),
    'solver' : ['liblinear'],
    'max_iter' : [10000,12500, 15000],
    'warm_start': [True, False]
    }    
logreg_opt = GridSearchCV(estimator = logreg, param_grid = logreg_params, cv = 10, n_jobs = -1,scoring = 'roc_auc')


rf = RandomForestClassifier()
rf_params={
    'max_depth': [2,3,4],
    'n_estimators': np.arange(50,300,50)}
    #'criterion': ['gini','entropy','log_loss']}
    #'max_features': np.arange(0.1,1,0.3),
    #'max_samples': [0.3,0,5,0.8]}              
rf_opt = GridSearchCV(estimator = rf, param_grid = rf_params, cv = 10, n_jobs = -1,scoring = 'f1')


bc = BaggingClassifier(n_jobs = -1)
bc_params = {
    #'max_features': np.arange(0.1,1,0.1),
    'max_samples' : np.linspace(0.5,1,10),
    'max_features' : np.linspace(0.5,1,10)}
    #'n_estimators' :np.arange(20,50,10)}
bc_opt = GridSearchCV(estimator = bc, param_grid = bc_params, cv = 10, n_jobs = -1,scoring = 'f1')


ag = AdaBoostClassifier()
ag_params = {
    'n_estimators' :np.arange(60,200,20),
    'learning_rate': [0.05, 0.1, 0.15,0.5,1] } 
ag_opt = GridSearchCV(estimator = ag, param_grid = ag_params, cv = 10, n_jobs = -1,scoring = 'f1')


sgbt = GradientBoostingClassifier()
sgbt_params = { 
    'max_depth': [2,3,4],
    'n_estimators' :np.arange(60,200,20),
    'subsample': np.arange(0.5,0.8,0.1) ,
    'max_features': np.arange(0.1,0.4,0.1)} 
sgbt_opt = GridSearchCV(estimator = sgbt, param_grid = sgbt_params, cv = 10, n_jobs = -1,scoring = 'f1')

```

```python

models_base = [svc, knn, dt, logreg, rf,bc,ag,sgbt]
weights =[]
models = [svc_opt, knn_opt, dt_opt, logreg_opt, rf_opt,bc_opt,ag_opt,sgbt_opt]
#models = [svc_opt, dt_opt, logreg_opt, rf_opt,bc_opt,ag_opt,sgbt_opt]
best_models = []
surv = pd.Series([0]*len(y_test.index), index = y_test.index)#pd.Series(dtype = int ,index = y_test.index)#[[],index = y_test.index]
for model in models:
    model.fit(X_train,y_train)
    best_models.append(model.best_estimator_)
    y_pred = model.predict(X_test)
    surv = surv.add(y)
    y_pred_train= model.predict(X_train)
    acc = accuracy_score(y_test,y_pred)
    acc_train = accuracy_score(y_train,y_pred_train)
    weights.append(acc)
    print(acc_train)
    print(acc)
    print('\n')
```
